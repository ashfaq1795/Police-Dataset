#### 5. What is mean of stop_duration?

```python
police_data.head()
```

here I have work only with column 'stop_duration' so I am finding its mean while it's in string form.

```python
#first i will check the different values in column.
police_data['stop_duration'].value_counts()
```

###### Graph for no of drivers (male and female separatly) vs violation

```python
x=police_data.groupby('violation').driver_gender.value_counts()
```

```python
x
```

```python
violation = ['Equipment', 'Moving violation', 'Other', 'Registration/plates', 'Seat belt', 'Speeding']
driver_gender_M = [5120, 9784, 3129, 1611, 3, 25517]
driver_gender_F = [1396, 2142, 454, 632, 0, 11686]

a = range(len(violation))
fig,ax = plt.subplots(figsize=(12,8))
bar_width = 0.25
opacity = 0.8

bar1 = ax.bar(a, driver_gender_M, bar_width, alpha=opacity, color='b', label='M')   #bar1
bar2 = ax.bar([i + bar_width for i in a], driver_gender_F, bar_width, alpha=opacity, color='r', label='F')  #bar2

ax.set_xlabel('Violation')
ax.set_ylabel('Number of Offenses')
ax.set_title('Violations by Driver Gender')
ax.set_xticks([i + bar_width / 2 for i in a], violation)
ax.legend()
fig.tight_layout()  #This line adjusts the layout of the chart so that all elements are properly spaced and aligned
```

From this graph it's cleared that in each violation the male gender has greater average than female.

```python
y=x.groupby(level=0).sum()   #level=0 represent previous attribute (violation) i,e it will groub by violation.
y
```
